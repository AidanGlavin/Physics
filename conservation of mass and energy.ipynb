# conservation of energy
## energy
### Gravitational energy
gravitational energy aka potential energy is the interacction of a body with the gravity of Earth or other large masses that have gravity

```python
import matplotlib.pyplot as plt
```

```python
def gravE(mass,height,gravity=9.8):
    '''gravE calculates the GPE given the mass in kg and the height in m, it uses earths gravity value of 9.8m/s by default'''
    return mass*height*gravity
```

```python
#list expansion
#masses= [n for n in range(1,100)]
masses=[]
for n in range(1,100):
    masses.append(n)
```

```python
#energies=[gravE(n, 100) for n in masses]

energies=[]
for n in masses:
    energies.append(gravE(n,100))
```

```python
plt.title("Energy in relation to mass at 100m")
plt.xlabel("Mass")
plt.ylabel("Energy in J")
plt.plot(masses,energies)
plt.show()
```

### comparing different values for G


```python
vesta_e=[]
for n in masses:
    vesta_e.append(gravE(n,100,gravity=0.3))
    
```

```python
mars_e=[]
for n in masses:
    mars_e.append(gravE(n,100,gravity=3.7))
```

```python
moon_e=[]
for n in masses:
    moon_e.append(gravE(n,100,gravity=1.6))
```

```python
jupiter_e=[]
for n in masses:
     jupiter_e.append(gravE(n,100,gravity=26))
```

```python
plt.title("comparing GPE on different celestial bodies")
plt.xlabel("mass")
plt.ylabel("energy in J")
plt.plot(masses,energies,label="earth")
plt.plot(masses,vesta_e,label="vesta")
plt.plot(masses,mars_e,label="Mars")
plt.plot(masses,moon_e,label="moon")
plt.plot(masses,jupiter_e,label="jupiter")
plt.legend()
plt.show()
```
