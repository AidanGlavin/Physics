# conservation of energy
## energy
### Gravitational energy
gravitational energy aka potential energy is the interacction of a body with the gravity of Earth or other large masses that have gravity

```python
import matplotlib.pyplot as plt
```

```python
def gravE(mass,height,gravity=9.8):
    return mass*height*gravity
```

```python
#list expansion
#masses= [n for n in range(1,100)]
masses=[]
for n in range(1,100):
    masses.append(n)
```

```python
#energies=[gravE(n, 100) for n in masses]

energies=[]
for n in masses:
    energies.append(gravE(n,100))
```

```python
plt.title("Energy in relation to mass at 100m")
plt.xlabel("Mass")
plt.ylabel("Energy in J")
plt.plot(masses,energies)
plt.show()
```
